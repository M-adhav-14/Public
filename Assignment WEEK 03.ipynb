## 5) Meaningful Visualisations

#### Histogram

```python
plt.figure()
plt.hist(data["SL"], color="red", density="true",rwidth=0.9)
plt.title("Data on SL")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
```

#### Scatter Diagram

```python
plt.figure(figsize=(6,4))
plt.scatter(data["SL"],data["PL"],s=15,c="red", marker="^")
plt.title("HP vs MPG", fontsize=10)
plt.xlabel("HP")
plt.ylabel("MPG")
```

#### Pair plots

```python
sns.pairplot(data)
```

#### Box Plots

```python
sns.boxplot(x="PW",y="SW", data=data)
```
