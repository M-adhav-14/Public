Check and handle outliers in at least 3 columns in the dataset

```python
# Handling outliers in Ticket Column.
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(y='Fare',data=titanic,palette='Accent')
```

```python
Q1=titanic.Fare.quantile(0.25)
Q3=titanic.Fare.quantile(0.75)
print(Q1,Q3)
```

```python
IQR=Q3-Q1
IQR
```

```python
lower=Q1-1.5*(IQR)
upper=Q3+1.5*(IQR)
print(lower,upper)
```

```python
outliers=titanic[(titanic.Fare<lower) | (titanic.Fare>upper)]
outliers
```

```python
titanic["Fare"] = np.where(titanic["Fare"]> upper, upper,
                        np.where(titanic["Fare"]< lower, lower,
                          titanic["Fare"]))
```

```python
titanic
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(y='Fare',data=titanic,palette='Accent')
```
